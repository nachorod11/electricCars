```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
url = r'https://api.openchargemap.io/v3/poi/?'
params = {"countrycode": "ES", "output": "json", "compact": True, "verbose": False, "maxresults": 5000,
         "key":"99dfe038-3369-4373-a05f-515e24bbfeca"}
data = requests.get(url, params = params)
```

```python
file = data.text
file = json.loads(file)
```

```python
# Guardamos los datos en un archivo JSON

with open('puntos_carga.json', 'w') as fout:
    json.dump(file , fout)
```
